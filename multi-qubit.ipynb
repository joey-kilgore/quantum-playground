```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
from qiskit.visualization import plot_bloch_multivector

from math import sqrt, pi

%matplotlib inline
from matplotlib import *
```

```python
qr = QuantumRegister(2)
cr = ClassicalRegister(2)
circuit = QuantumCircuit(qr, cr)
circuit.h(qr)
circuit.measure(qr, cr)
circuit.draw(output='mpl', scale=2.5)
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator).result()
plot_histogram(result.get_counts(circuit))
```

# Entanglement
We will use the CNOT-gate (cx) to entangle two bits together. the CNOT gate is a controlled X-gate based on one of the qubits and performs the action on the second qubit. 

| input(t,c) | Output(t,c) |
| --- | --- |
| 00 | 00 |
| 01 | 11 |
| 10 | 10 |
| 11 | 01 |

```python
circuit = QuantumCircuit(qr, cr)
circuit.h(qr[0])
circuit.cx(qr[0],qr[1])
circuit.measure(qr, cr)
circuit.draw(output='mpl', scale=2.5)
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator).result()
plot_histogram(result.get_counts(circuit))
```

```python
qr = QuantumRegister(3)
cr = ClassicalRegister(3)
circuit = QuantumCircuit(qr, cr)
circuit.h(qr[0])
circuit.h(qr[1])
circuit.ccx(qr[0], qr[1], qr[2])
circuit.ccx(qr[2], qr[1], qr[0])
circuit.measure(qr, cr)
circuit.draw(output='mpl', scale=2.5)
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator).result()
plot_histogram(result.get_counts(circuit))
```
